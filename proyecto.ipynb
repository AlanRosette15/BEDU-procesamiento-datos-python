



Para sustituir los valores categoricos en las columnas con terminacion _ID por valores más descriptivos, hacemos un mapeo de sus valores en otros dataframes para después sustituirlos en un dataframe final

```
clear_dict = dict(df_clear[["Clear_ID", "ClearName"]].values)
colores_dict = dict(df_colores[["Color_ID", "ColorName"]].values)
primers_dict = dict(df_primers[["Primer_ID", "PrimerName"]].values)
product_dict = dict(df_product[["Product_ID", "ProductName"]].values)
types_dict = dict(df_types[["Job_ID", "JobName"]].values)
types_dict[24] = 'Tigua_PA'

df_consumo = df.copy()
df_consumo["Clear_ID"] = df_consumo["Clear_ID"].map(clear_dict)
df_consumo["Color_ID"] = df_consumo["Color_ID"].map(colores_dict)
df_consumo["Primer_ID"] = df_consumo["Primer_ID"].map(primers_dict)
df_consumo["Product_ID"] = df_consumo["Product_ID"].map(product_dict)
df_consumo["Job_ID"] = df_consumo["Job_ID"].map(types_dict)



```

El nuevo dataframe llamado `df_consumo` es una copia del dataframe original, pero con los valores _ID actualizados. Falta reemplazar los valores `NaN` por `0`


```
df_consumo = df_consumo.fillna(0)
df_consumo
```

```
def calcular_consumo(robots):
  """Calcula el consumo de pintura de un par de robots.

  Parametros
  ----------
  robots : pd.Series
    Una serie con la cantidad de consumo de pintura de cada robot.
  
  Retorna
  -------
  float
    Consumo calculado de los dos robots.
  """
  if robots[0] == 0 or robots[1] == 0:
    return robots.sum() / 2
  else:
    return robots.sum() / 4
```

Se obtiene el consumo de material de cada par de robots

```
pares_robots = zip(df_consumo.iloc[:, range(8, 27, 2)].keys(), df_consumo.iloc[:, range(9, 28, 2)].keys())

for robot1, robot2 in pares_robots:
  nueva_columna = '_and_'.join([robot1,robot2])
  df_consumo[nueva_columna] = df_consumo[[robot1, robot2]].apply(lambda row: calcular_consumo(row), axis=1)
```

```
df_consumo.head()
```

Se agrega una columna con el consumo total de los 3 pares de robots de base

```
df_consumo['Resin_BC_Total'] = df_consumo.loc[:, 'Resin_R1_BC_and_Resin_R2_BC': 'Resin_R5_BC_and_Resin_R6_BC'].sum(axis=1)
```

```
df_consumo.head()
```

Para poder evaluar los consumos semanalmente, agregamos una nueva columna. Con ayuda de la función isocalendar, podemos obtener el numero de la semana del año a partir de la Fecha.

```
df_consumo['Semana']=df_consumo.Fecha.dt.isocalendar().week
```

Agrupamos los consumos por la semana, tipo de pieza (Job_ID), color y renombramos las columnas para mejor entendimiento. Es importante notar que no tomamos en cuenta los pares de robots (Resin CC_R1 y CC_R1, así como Hardener CC_R1 y CC_R2 ), debido a estándares de la empresa

```

df_consumo_semanal=df_consumo.groupby(['Semana',
                                       'Job_ID',
                                       'Color_ID'],as_index=False)[
                                      ['Resin_R1_PR_and_Resin_R2_PR',
                                      'Hardener_R1_PR_and_Hardener_R2_PR',
                                      'Resin_BC_Total',
                                      'Resin_R3_CC_and_Resin_R4_CC',
                                      'Hardener_R3_CC_and_Hardener_R4_CC']
                                       ].mean().rename(columns={
                                      'Resin_R1_PR_and_Resin_R2_PR':'Resina_Primer',
                                      'Hardener_R1_PR_and_Hardener_R2_PR':'Hardener_Primer',
                                      'Resin_BC_Total':'Color',
                                      'Resin_R3_CC_and_Resin_R4_CC':'Resina_ClearCoat',
                                      'Hardener_R3_CC_and_Hardener_R4_CC':'Hardener_ClearCoat'
                                      })

df_consumo_semanal.head
```
